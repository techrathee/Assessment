### Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
Answer.
Seaborn is a popular data visualization library in Python that builds on top of Matplotlib. It provides a high-level interface 
for creating attractive and informative statistical graphics. Here are five common plots that can be created using Seaborn, 
along with their uses:
    
    1. Scatter Plot: A scatter plot represents the relationship between two variables by displaying individual data points as
    markers on a two-dimensional plane. It is useful for exploring the correlation or pattern between two continuous variables.
    
    2. Line Plot: A line plot displays data points connected by straight lines, which is useful for showing trends or changes 
    over time. It is commonly used for time series analysis or to visualize the progression of a variable.
    
    3. Bar Plot: A bar plot presents categorical data with rectangular bars, where the height of each bar represents the value 
    of the category. It is suitable for comparing the values of different categories or displaying the distribution of a 
    categorical variable.
    
    4. Histogram: A histogram is a graphical representation of the distribution of a continuous variable. It divides the data 
    into bins and displays the frequency or count of observations within each bin. Histograms are useful for understanding the 
    shape, central tendency, and spread of a dataset.
    
    5. Heatmap: A heatmap uses color intensity to represent the values of a two-dimensional dataset in a tabular form. 
    It is often employed to visualize the correlation or relationship between multiple variables. Heatmaps are commonly used 
    in exploratory data analysis and in identifying patterns in complex datasets.
    
```

### Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
#Answer.
# Here's an example
import seaborn as sns

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Create a line plot using Seaborn
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
```

### Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
* Note: pclass, age, and fare are columns in the titanic dataset.

```python
#Answer.
#Here's an example
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Create the first box plot for 'age'
plt.subplot(1, 2, 1)  # Create a subplot for the first box plot
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Box Plot of Age')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

# Create the second box plot for 'fare'
plt.subplot(1, 2, 2)  # Create a subplot for the second box plot
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Box Plot of Fare')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

# Adjust the layout to prevent overlap of subplots
plt.tight_layout()

# Display the plot
plt.show()
```

### Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
#Answer.
#Here's an example
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Plot histogram with hue
sns.histplot(data=diamonds, x="price", hue="cut", kde=True)

# Set plot title and labels
plt.title("Histogram of Diamond Prices")
plt.xlabel("Price")
plt.ylabel("Count")

# Display the plot
plt.show()
```

### Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
