### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

```python
Answer.
 Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. 
    It provides a wide range of plotting capabilities and is extensively used for data exploration, analysis, and presentation.

Matplotlib is used for several reasons:
    
    1. Data Visualization: Matplotlib allows users to create various types of plots and charts to represent data in a visually
    appealing manner. It provides a high level of customization, enabling users to control almost every aspect of the plot.
    
    2. Data Exploration and Analysis: Matplotlib offers a wide range of plotting functions and tools, making it useful for exploring 
    and analyzing data. It allows users to create histograms, scatter plots, line plots, bar charts, and more to gain insights from
    data.
    
    3. Presentation and Reporting: Matplotlib provides a flexible platform to create publication-quality figures and plots.
    It allows users to customize the appearance of plots, add labels, legends, titles, annotations, and export plots in various 
    file formats for use in presentations, reports, or publications.
    
    4. Integration with Other Libraries: Matplotlib integrates well with other scientific computing libraries in Python, such as
    NumPy and pandas. It can directly plot data from these libraries and work seamlessly with them.
    
    5. Interactive Visualization: Matplotlib can be used in conjunction with other libraries, such as Jupyter Notebook or JupyterLab,
    to create interactive visualizations. These interactive plots can be manipulated and explored by users, enhancing the
    understanding of data.

Five plots that can be plotted using the Pyplot module of Matplotlib are:
    
    1. Line Plot: A basic plot that represents data points connected by straight lines. It is useful for showing the trend or 
    relationship between two variables over a continuous domain.
    
    2. Scatter Plot: A plot that represents individual data points as dots on a two-dimensional coordinate system. It is helpful 
    for visualizing the distribution and relationship between two variables.
    
    3. Bar Chart: A chart that represents data with rectangular bars, where the length of each bar corresponds to the value of the 
    data. It is commonly used for comparing categorical data or showing counts or frequencies.
    
    4. Histogram: A graphical representation of the distribution of numerical data. It divides the data into bins or intervals and 
    shows the frequency or proportion of data points falling into each bin.
    
    5. Pie Chart: A circular statistical chart that divides a circle into sectors, each representing a proportionate part of a whole.
    It is commonly used to display the composition or distribution of categorical data.
        
```

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
* import numpy as np
* np.random.seed(3)
* x = 3 + np.random.normal(0, 2, 50)
* y = 3 + np.random.normal(0, 2, len(x))
##### Note: Also add title, xlabel, and ylabel to the plot.

```python
#Answer.
# Here's the code:
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)  # Create the scatter plot
plt.title('Scatter Plot')  # Add a title to the plot
plt.xlabel('X')  # Add label for x-axis
plt.ylabel('Y')  # Add label for y-axis

plt.show()  # Display the plot
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
#### Use the following data:
* import numpy as np
* For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
* For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
* For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
* For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
#Answer.
# Here's the code:
import matplotlib.pyplot as plt
import numpy as np

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
* import numpy as np
* company = np.array(["Apple", "Microsoft", "Google", "AMD"])
* profit = np.array([3000, 8000, 1000, 10000])

```python
Answer.
A bar plot, also known as a bar chart or bar graph, is a type of plot that uses rectangular bars to represent categorical data.
The length of each bar corresponds to the magnitude or value of the data being represented. Bar plots are commonly used to compare
and visualize categorical data and their associated values.

Bar plots are used for several reasons:
    
    1. Comparison: Bar plots make it easy to compare the values of different categories. By visually comparing the lengths of the
    bars, you can quickly identify which categories have higher or lower values.
    
    2. Categorical Data Representation: Bar plots are effective for representing categorical data and their corresponding values. 
    Each category is represented by a distinct bar, making it clear which values belong to which categories.
    
    3. Presentation of Frequencies or Counts: Bar plots are often used to display the frequency or count of different categories.
    This is useful for visualizing distributions or showing the occurrence of specific events or categories.
    
    4. Trend Analysis: Bar plots can also be used to analyze trends over time or across different groups. By representing data in a
    categorical manner, you can identify patterns or changes in values between different categories.
    
```

```python
#Now, let's use the provided data to plot a bar plot and a horizontal bar plot using Matplotlib:
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.title('Profit Comparison')
plt.xlabel('Company')
plt.ylabel('Profit')

plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title('Profit Comparison')
plt.xlabel('Profit')
plt.ylabel('Company')

plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
* box1 = np.random.normal(100, 10, 200)
* box2 = np.random.normal(90, 20, 200)

```python
Answer.

A box plot, also known as a box-and-whisker plot, is a statistical visualization that displays the distribution of a dataset through 
quartiles. It summarizes the minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile),
and maximum values of the dataset, as well as any potential outliers.

Box plots are used for several reasons:
    
    1. Data Distribution: Box plots provide a concise and visual summary of the distribution of a dataset. They show the spread, 
    skewness, and symmetry of the data, allowing you to identify the presence of outliers or unusual values.
    
    2. Comparison: Box plots are effective for comparing the distributions of multiple datasets side by side. By placing multiple
    box plots on the same axis, you can easily compare the medians, quartiles, and spread of different datasets.
    
    3. Identification of Outliers: Box plots visually represent outliers, which are data points that are significantly different
    from the majority of the data. Outliers can be identified as individual points plotted outside the whiskers of the box plot.
    
    4. Data Analysis: Box plots are commonly used in statistical analysis to assess the distribution of data, detect skewness or 
    asymmetry, and identify any unusual or extreme values. They provide a comprehensive overview of the dataset's central tendency
    and spread.
    
```

```python
#Now, let's use the provided data to plot a box plot using Matplotlib:
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Values')

plt.show()
```
