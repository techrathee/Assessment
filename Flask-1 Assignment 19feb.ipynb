### Q1. What is Flask Framework? What are the advantages of Flask Framework?

```python
Answer.
Flask is a popular micro web framework written in Python. It is designed to be lightweight, flexible, and easy to use, making it a popular
choice for developing web applications and APIs.

Here are some advantages of using the Flask framework:
    
    1. Lightweight and minimalistic: Flask is a micro framework, which means it has a small and simple core that can be easily extended
    with additional libraries as needed. This minimalistic approach allows developers to have more control over their applications and
    only include the features they actually need.
    
    2. Easy to get started: Flask has a straightforward and intuitive API that makes it easy for beginners to learn and start building 
    web applications. The framework provides a solid foundation for web development, handling routing, HTTP requests, and responses,
    while allowing developers to focus on application logic.
    
    3. Flexible and customizable: Flask gives developers the freedom to design and structure their applications according to their specific
    needs. It does not impose strict architectural patterns, allowing you to choose the components and extensions that best suit your
    project requirements.
    
    4. Extensible ecosystem: Flask has a large and active community, which has contributed a wide range of extensions and plugins.
    These extensions provide additional functionality, such as database integration, authentication, form validation, and more. 
    The Flask ecosystem offers numerous options to extend and enhance your application with ease.
    
    5. Well-documented: Flask has comprehensive and well-organized documentation that covers all aspects of the framework.
    The documentation includes detailed examples, tutorials, and explanations of various concepts, making it easier for developers
    to learn and use Flask effectively.
    
```

### Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
Answer.
Here's an example of a Flask application that displays "Hello World!!":
 
    from flask import Flask

    app = Flask(__name__)

    @app.route("/")
    def hello_world():
        return "<h1>Hello, World!!</h1>"

    if __name__=="__main__":
        app.run(host="0.0.0.0")
        
```

### Q3. What is App routing in Flask? Why do we use app routes?

```python
Answer.
In Flask, app routing refers to the process of mapping URLs (Uniform Resource Locators) to specific functions in your application.
It allows you to define different routes or endpoints that the application can handle, based on the URL requested by the client.

When a client makes a request to a Flask application, the application needs to determine which code should be executed to handle that request.
This is where app routing comes into play. By using app routes, you can specify which functions should be called for specific URLs or URL patterns.

Here's an example of defining an app route in Flask:
 
    from flask import Flask

    app = Flask(__name__)

    @app.route('/')
    def home():
        return 'This is the home page.'

    @app.route('/about')
    def about():
        return 'This is the about page.'

    if __name__=="__main__":
        app.run(host="0.0.0.0")
```

### Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
* Company Name: ABC Corporation
* Location: India
* Contact Detail: 999-999-9999

`Attach the screenshot of the output in Jupyter Notebook.`

```python
Answer.
 Here's an example of a Flask application that includes the /welcome and root (/) routes you described:
    
    from flask import Flask

    app = Flask(__name__)

    @app.route('/welcome')
    def welcome():
        return 'Welcome to ABC Corporation'

    @app.route('/')
    def company_details():
        return '''<h1>Company Name: ABC Corporation
            Location: India
            Contact Detail: 999-999-9999</h1>'''

    if __name__ == '__main__':
        app.run(host="0.0.0.0")
        
```
