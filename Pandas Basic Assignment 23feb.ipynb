### Q1. Create a Pandas Series that contains the following data: 4, 8, 15, 16, 23, and 42. Then, print the series.

### Q3. Create a Pandas DataFrame that contains the following data:
`Name     Age    Gender`
`Alice    25     Female`
`Bob      30     Male`
`Claire   27     Female`
#### Then, print the DataFrame.

### Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

```python
Answer.
In pandas, a DataFrame is a two-dimensional labeled data structure that consists of columns and rows. It is similar to a table in a relational database or a spreadsheet.
Each column in a DataFrame represents a different variable or feature, while each row represents an observation or record.

A DataFrame can be thought of as a collection of Pandas Series objects. Each column in a DataFrame is a Pandas Series, and all the columns together form the DataFrame.
The key difference between a DataFrame and a Series is that a DataFrame can store and manipulate multiple columns of data, while a Series can only store a single column of data.
```

```python
#Here's an example to illustrate the difference between a DataFrame and a Series:
import pandas as pd

# Creating a Series
series = pd.Series([4, 8, 15, 16, 23, 42])
print(series)
```

```python
In the code above, we create a Pandas Series object called series with the numbers 4, 8, 15, 16, 23, and 42. When we print the series, it displays the values along with their
corresponding indices.
```

```python
#Now let's create a DataFrame using the same data:
import pandas as pd

# Creating a DataFrame
data = {
    'Numbers': [4, 8, 15, 16, 23, 42]
}
df = pd.DataFrame(data)
print(df)
```

### Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

```python
Answer.
Here are some common functions:
    1. head() and tail():
        These functions allow you to view the first few rows (head()) or last few rows (tail()) of a DataFrame. They are useful for quickly inspecting the data.
    
```

```python
#Example:
#Head
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head()
```

```python
#tail
df.tail()
```

```python
2. groupby():
    This function enables you to group the data based on one or more columns and perform aggregate operations on the grouped data.
    It is helpful for generating summary statistics for different groups in the data.
```

```python
#Example 
g = df.groupby("Pclass")
g.sum()
```

```python
3. sort_values():
    This function allows you to sort the DataFrame by one or more columns. You can specify the column(s) to sort by and the sorting order (ascending or descending).
```

```python
df.sort_values('Age', ascending=False)
```

```python
4. describe():
    This function computes various summary statistics of the numerical columns in the DataFrame, such as count, mean, standard deviation, minimum, and maximum.
    It helps in understanding the distribution of the data.
    
```

```python
df.describe()
```

```python
5. info():
    This function provides a summary of the DataFrame, including the data types of each column and the number of non-null values. It is useful for getting an overview of the
    DataFrame's structure.
    
```

```python
df.info()
```

### Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

```python
Answer.
Among the options given (Series, DataFrame, Panel), both Series and DataFrame are mutable in nature, while Panel is not.

In Pandas, a mutable object can be modified or changed after it is created. Let's look at each option':
    
    1. Series: A Pandas Series is mutable, which means you can modify its values, add new elements, or change existing elements.
    
    2. DataFrame: Similarly, a Pandas DataFrame is also mutable. You can modify the values in individual cells, add or remove columns, or perform various operations to manipulate the 
    data within the DataFrame.
    
    3. Panel: However, Panel is not mutable. Panel objects have been deprecated since Pandas version 0.20. Instead, it is recommended to use multi-dimensional NumPy arrays or
    a combination of DataFrames for handling three-dimensional data.
    
```

### Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
#Answer
#Here's an example:
import pandas as pd

# Creating Series
name_series = pd.Series(['Alice', 'Bob', 'Claire'])
age_series = pd.Series([25, 30, 27])
gender_series = pd.Series(['Female', 'Male', 'Female'])

# Creating DataFrame
data = {
    'Name': name_series,
    'Age': age_series,
    'Gender': gender_series
}
df = pd.DataFrame(data)

# Printing DataFrame
print(df)
```
